```python
from mlconjug3 import Conjugator
```

```python
# initialize the conjugator
conjugator = Conjugator(language="pt")

# conjugate the verb "parler"
verb = conjugator.conjugate("caber")

# print all the conjugated forms as a list of tuples.
verb.iterate()
```
